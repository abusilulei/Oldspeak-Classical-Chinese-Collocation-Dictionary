# Connect to the Database

```python
import pandas as pd
import psycopg2
from psycopg2.extras import execute_values
import json
from tqdm import tqdm  # For progress tracking

# read the password from the .env file
from dotenv import load_dotenv
import os
load_dotenv()
db_url = os.getenv("DATABASE_URL")
print(f"Connecting to database at {db_url}")
if db_url is None:
    raise ValueError("DATABASE_URL not found in .env file")
```

# Insert Data to the Database

```python
import os
import tempfile
import csv
from io import StringIO

# Configuration
BATCH_SIZE = 10000  # Adjust based on memory constraints

def prepare_csv_files(books_df, linguistic_df):
    print("Preparing CSV files for bulk import...")
    
    # Create a temporary directory for CSV files
    temp_dir = "./temp"
    print(f"Using temporary directory: {temp_dir}")
    
    # 1. Process books
    print("Processing books...")
    books_csv_path = os.path.join(temp_dir, "books.csv")
    books_df = books_df[['name', 'type', 'time', 'style']].drop_duplicates()
    books_df.to_csv(books_csv_path, index=False, header=False, quoting=csv.QUOTE_MINIMAL)
    
    # 2. Prepare book ID mapping
    book_id_map = {}
    next_book_id = 1
    for book_name in books_df['name'].unique():
        book_id_map[book_name] = next_book_id
        next_book_id += 1
    
    # 3. Process sentences
    print("Processing sentences...")
    sentences_csv_path = os.path.join(temp_dir, "sentences.csv")
    sentence_id_map = {}
    next_sentence_id = 1
    
    with open(sentences_csv_path, 'w', newline='') as f_sentences:
        sentences_writer = csv.writer(f_sentences, quoting=csv.QUOTE_MINIMAL)
        
        # Group by book and deduplicate sentences
        processed_sentences = set()
        for _, row in tqdm(linguistic_df.iterrows(), total=len(linguistic_df)):
            book_id = book_id_map.get(row['name'])
            if not book_id:
                continue
                
            sentence_text = row['sentences']
            sentence_key = f"{book_id}:{sentence_text}"
            
            if sentence_key not in processed_sentences:
                processed_sentences.add(sentence_key)
                sentences_writer.writerow([next_sentence_id, book_id, sentence_text])
                sentence_id_map[sentence_key] = next_sentence_id
                next_sentence_id += 1
    
    # 4. Process tokens
    print("Processing tokens...")
    tokens_csv_path = os.path.join(temp_dir, "tokens.csv")
    token_id_map = {}
    next_token_id = 1
    
    with open(tokens_csv_path, 'w', newline='') as f_tokens:
        tokens_writer = csv.writer(f_tokens, quoting=csv.QUOTE_MINIMAL)
        
        for _, row in tqdm(linguistic_df.iterrows(), total=len(linguistic_df)):
            book_id = book_id_map.get(row['name'])
            if not book_id:
                continue
                
            sentence_text = row['sentences']
            sentence_key = f"{book_id}:{sentence_text}"
            sentence_id = sentence_id_map.get(sentence_key)
            
            if not sentence_id:
                continue
            
            for pos, (token, pos_tag) in enumerate(zip(row['token'], row['upos'])):
                tokens_writer.writerow([next_token_id, sentence_id, pos, token, pos_tag])
                token_key = f"{sentence_id}:{pos}"
                token_id_map[token_key] = next_token_id
                next_token_id += 1
    
    # 5. Process dependencies
    print("Processing dependencies...")
    dependencies_csv_path = os.path.join(temp_dir, "dependencies.csv")
    next_dep_id = 1
    
    with open(dependencies_csv_path, 'w', newline='') as f_deps:
        deps_writer = csv.writer(f_deps, quoting=csv.QUOTE_MINIMAL)
        
        for _, row in tqdm(linguistic_df.iterrows(), total=len(linguistic_df)):
            book_id = book_id_map.get(row['name'])
            if not book_id:
                continue
                
            sentence_text = row['sentences']
            sentence_key = f"{book_id}:{sentence_text}"
            sentence_id = sentence_id_map.get(sentence_key)
            
            if not sentence_id:
                continue
            
            for pos, (head_pos, dep_type) in enumerate(row['dep']):
                dependent_key = f"{sentence_id}:{pos}"
                dependent_id = token_id_map.get(dependent_key)
                
                if head_pos == 0:  # Root
                    head_id = dependent_id  # Self-reference
                else:
                    head_key = f"{sentence_id}:{head_pos - 1}"
                    head_id = token_id_map.get(head_key)
                
                if dependent_id and head_id:
                    deps_writer.writerow([next_dep_id, head_id, dependent_id, dep_type])
                    next_dep_id += 1
    
    return {
        'books_csv': books_csv_path,
        'sentences_csv': sentences_csv_path,
        'tokens_csv': tokens_csv_path,
        'dependencies_csv': dependencies_csv_path
    }

def import_csv_to_postgres_robust(csv_paths):
    print("Importing CSV files to PostgreSQL...")
    conn = psycopg2.connect(db_url)
    
    try:
        with conn.cursor() as cur:
            # Settings for faster import
            cur.execute("SET session_replication_role = 'replica';")
            cur.execute("SET client_encoding = 'UTF8';")
            cur.execute("SET maintenance_work_mem = '256MB';")  # For faster constraint processing later
            
            # Clear tables
            print("Clearing existing data...")
            cur.execute("TRUNCATE dependencies, tokens, sentences, books RESTART IDENTITY CASCADE;")
            
            # Use psycopg2's copy_expert for more control
            for table, file_path, columns in [
                ('books', csv_paths['books_csv'], ['name', 'category', 'period', 'style']),
                ('sentences', csv_paths['sentences_csv'], ['sentence_id', 'book_id', 'content']),
                ('tokens', csv_paths['tokens_csv'], ['token_id', 'sentence_id', 'position', 'token_text', 'pos_tag']),
                ('dependencies', csv_paths['dependencies_csv'], ['dependency_id', 'head_token_id', 'dependent_token_id', 'dependency_type'])
            ]:
                print(f"Importing {table}...")
                with open(file_path, 'r') as f:
                    columns_str = ', '.join(columns)
                    sql = f"COPY {table} ({columns_str}) FROM STDIN WITH CSV"
                    cur.copy_expert(sql=sql, file=f)
            
            # Reset sequences
            print("Updating sequence values...")
            tables = ['books', 'sentences', 'tokens', 'dependencies']
            id_cols = ['book_id', 'sentence_id', 'token_id', 'dependency_id']
            
            for table, id_col in zip(tables, id_cols):
                cur.execute(f"SELECT setval('{table}_{id_col}_seq', (SELECT MAX({id_col}) FROM {table}), true);")
            
        conn.commit()
        print("Import completed successfully!")
        
    except Exception as e:
        conn.rollback()
        print(f"Error during import: {e}")
        raise
    finally:
        conn.close()

def fast_import(books_df, linguistic_df):
    try:
        # Prepare CSV files
        csv_paths = prepare_csv_files(books_df, linguistic_df)
        
        # Import to PostgreSQL
        import_csv_to_postgres_robust(csv_paths)
        
        # Clean up temporary CSV files
        for csv_path in csv_paths.values():
            if os.path.exists(csv_path):
                os.remove(csv_path)
                
    except Exception as e:
        print(f"Error during fast import: {e}")
        raise

# Execute the fast import
fast_import(books_info, tok_pos_dep)
```
